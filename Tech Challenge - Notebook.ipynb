# Hypothesis Testing

- **H0**: The fuel consumption of E10 and SP98 are the same.
- **H1** : The fuel consumption of E10 and SP98 are NOT the same.

```python
import scipy.stats as stats

E10 = df[df['gas_type'] == "E10"]["fuel_usage"]
SP98 = df[df['gas_type'] == "SP98"]["fuel_usage"]

t_stat, p_value = stats.ttest_ind(E10, SP98)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (H0). The fuel consumption is different.")
else:
    print("Fail to reject the null hypothesis (H0). The fuel consumption is not significantly different.")

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

#### We accept our null hypothesis, the p-value is above 0.05

# Calculating costs per trip

```python
#Prices in Barcelona
E10_price = 1.79 
SP98_price = 1.91  

# Calculating consumption per trip in liters
df['consume_per_trip'] = (df['consume'] / 100) * df['distance']

# Calculating cost per trip based on current gas prices
def add_trip_cost(row):
    consumption = row['consume_per_trip']
    if row["gas_type"] == "E10":
        trip_cost = consumption * E10_price
    else: 
        trip_cost = consumption * SP98_price
    return trip_cost

df['trip_cost'] = df.apply(add_trip_cost, axis=1)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lmplot(x='speed', y='trip_cost', hue='gas_type', data=df, aspect=2, height=6)
plt.title('Cost of Trip vs. Speed for E10 and SP98')
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Trip Cost (€)')

plt.show()
```

**Plot of Trip Cost vs. Speed**:
The plot indicates that, when traveling at high speeds, SP98 generally results in higher trip costs compared to E10. Nonetheless, it's crucial to emphasize that the available data is not extensive enough to firmly substantiate this observation.
